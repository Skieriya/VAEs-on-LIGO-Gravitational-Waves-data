```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Physics Simulation
def generate_noisy_bh_signal(n_samples=1000, noise_level=0.3):
    t = torch.linspace(0, 5, 100)
    noisy_signals, clean_signals, parameters = [], [], []
    for _ in range(n_samples):
        mass = torch.rand(1) * 2.0 + 0.5
        spin = torch.rand(1) * 0.5 + 0.1
        clean = torch.exp(-t * spin) * torch.cos(mass * 10 * t)
        noise = torch.randn_like(clean) * noise_level
        noisy_signals.append(clean + noise)
        clean_signals.append(clean)
        parameters.append(torch.tensor([mass, spin]))
    return torch.stack(noisy_signals), torch.stack(clean_signals), torch.stack(parameters)

# 2. Multi-Task VAE Model
class PhysicsInferenceVAE(nn.Module):
    def __init__(self, signal_dim=100, latent_dim=8):
        super(PhysicsInferenceVAE, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(signal_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, latent_dim * 2)
        )
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, signal_dim)
        )
        self.regressor = nn.Sequential(
            nn.Linear(latent_dim, 32),
            nn.ReLU(),
            nn.Linear(32, 2)
        )

    def reparameterize(self, mu, log_var):
        std = torch.exp(0.5 * log_var)
        eps = torch.randn_like(std)
        return mu + eps * std

    def forward(self, x_noisy):
        h = self.encoder(x_noisy)
        mu, log_var = torch.chunk(h, 2, dim=1)
        z = self.reparameterize(mu, log_var)
        recon_clean = self.decoder(z)
        pred_params = self.regressor(mu)
        return recon_clean, pred_params, mu, log_var

# 3. Metrics Calculation (MAPE)
def calculate_mape(actual, predicted):
    return torch.mean(torch.abs((actual - predicted) / actual)) * 100

# 4. Training and Benchmark Logic
noisy_train, clean_train, true_params = generate_noisy_bh_signal(5000)
model = PhysicsInferenceVAE()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

print("Starting Training...")
for epoch in range(1201):
    model.train()
    optimizer.zero_grad()
    recon, pred_p, mu, log_var = model(noisy_train)

    recon_loss = nn.functional.mse_loss(recon, clean_train, reduction='sum')
    kld_loss = -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp())
    reg_loss = nn.functional.mse_loss(pred_p, true_params, reduction='sum') * 50

    loss = recon_loss + (0.1 * kld_loss) + reg_loss
    loss.backward()
    optimizer.step()

    if epoch % 200 == 0:
        mape = calculate_mape(true_params, pred_p)
        print(f"Epoch {epoch} | Loss: {loss.item():.2f} | Physics MAPE: {mape.item():.2f}%")

# 5. Baseline Comparison (Standard Linear Regression)
print("\nCalculating Baseline Benchmark...")
baseline_model = LinearRegression()
baseline_model.fit(noisy_train.numpy(), true_params.numpy())

# 6. Evaluation
model.eval()
with torch.no_grad():
    test_noisy, test_clean, test_actual = generate_noisy_bh_signal(n_samples=500, noise_level=0.4)

    # VAE Predictions
    _, vae_preds, _, _ = model(test_noisy)
    vae_mape = calculate_mape(test_actual, vae_preds)

    # Baseline Predictions
    base_preds = torch.tensor(baseline_model.predict(test_noisy.numpy()))
    base_mape = calculate_mape(test_actual, base_preds)

    print("-" * 40)
    print(f"BENCHMARK RESULTS (Noise Level 0.4)")
    print(f"Linear Regression MAPE: {base_mape.item():.2f}%")
    print(f"VAE Inference MAPE:    {vae_mape.item():.2f}%")
    print(f"Improvement:           {base_mape.item() - vae_mape.item():.2f}% points")
    print("-" * 40)

# 7. Visualization
    sample_noisy, sample_clean, sample_actual = generate_noisy_bh_signal(n_samples=1, noise_level=0.4)
    denoised, predicted_physics, _, _ = model(sample_noisy)

    plt.figure(figsize=(10, 5))
    plt.plot(sample_noisy[0], label="Noisy Observation", alpha=0.3, color='red')
    plt.plot(sample_clean[0], label="Ground Truth", linestyle='--', color='green')
    plt.plot(denoised[0], label="VAE Denoised Output", color='blue', linewidth=2)
    plt.title(f"Target Mass: {sample_actual[0,0]:.2f} | VAE Predicted: {predicted_physics[0,0]:.2f}")
    plt.legend()
    plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import numpy as np
from gwpy.timeseries import TimeSeries
from gwosc.datasets import event_gps
from pycbc.waveform import get_td_waveform
import os
import time



def generate_realistic_synthetic(n_samples=1000):
    print(f"Generating {n_samples} synthetic waveforms with the reliable 'IMRPhenomD' model...")
    noisy, params = [], []
    sample_rate = 1024
    t_target = np.linspace(-0.02, 0.02, 100)
    approximant = 'IMRPhenomD'

    for i in range(n_samples):
        if (i+1) % 100 == 0:
            print(f"  ... generated {i+1}/{n_samples} waveforms")

        mass1 = torch.rand(1) * 20.0 + 20.0
        mass2 = mass1
        spin = torch.rand(1) * 1.6 - 0.8

        hp, _ = get_td_waveform(approximant=approximant,
                                 mass1=mass1.item(), mass2=mass2.item(),
                                 spin1z=spin.item(), delta_t=1.0/sample_rate, f_lower=20)

        merger_time = hp.sample_times[np.argmax(abs(hp.data))]
        hp.start_time = -merger_time
        clean_sig = np.interp(t_target, hp.sample_times, hp.data)
        clean_sig = torch.from_numpy(clean_sig).float()

        noise = torch.randn_like(clean_sig) * 0.5
        noisy_sample = clean_sig + noise
        noisy_sample = (noisy_sample - noisy_sample.mean()) / noisy_sample.std()

        noisy.append(noisy_sample)
        total_mass = mass1 + mass2
        params.append(torch.tensor([total_mass.item(), spin.item()]))

    return torch.stack(noisy), torch.stack(params)


class LIGO_VAE(nn.Module):
    def __init__(self, signal_dim=100, latent_dim=16):
        super(LIGO_VAE, self).__init__()
        self.encoder = nn.Sequential(nn.Linear(signal_dim, 256), nn.ReLU(), nn.Linear(256, 128), nn.ReLU(), nn.Linear(128, latent_dim * 2))
        self.decoder = nn.Sequential(nn.Linear(latent_dim, 128), nn.ReLU(), nn.Linear(128, 256), nn.ReLU(), nn.Linear(256, signal_dim))
        self.regressor = nn.Sequential(nn.Linear(latent_dim, 64), nn.ReLU(), nn.Linear(64, 2))
    def reparameterize(self, mu, log_var):
        std = torch.exp(0.5 * log_var); eps = torch.randn_like(std); return mu + eps * std
    def forward(self, x):
        h = self.encoder(x); mu, log_var = torch.chunk(h, 2, dim=1)
        z = self.reparameterize(mu, log_var)
        return self.decoder(z), self.regressor(z), mu, log_var

def get_real_few_shot_dataset():
    events = {'GW150914': {'mass': 65.3, 'spin': -0.06}, 'GW151226': {'mass': 21.8, 'spin': 0.21}, 'GW170104': {'mass': 60.6, 'spin': -0.12}, 'GW170814': {'mass': 56.1, 'spin': 0.06}, 'GW170608': {'mass': 35.1, 'spin': 0.07}}
    real_data, real_params = [], []
    print("Fetching and processing real events for fine-tuning...")
    for event_name, params in events.items():
        try:
            t0 = event_gps(event_name); raw = TimeSeries.fetch_open_data('H1', t0-1, t0+1, cache=True)
            clean = raw.whiten().bandpass(30, 400); cropped = clean.crop(t0-0.02, t0+0.02).value[:100].copy()
            test_input = torch.from_numpy(cropped).float(); test_input = (test_input - test_input.mean()) / test_input.std()
            real_data.append(test_input); real_params.append(torch.tensor([params['mass'], params['spin']]))
            print(f"  - Successfully processed {event_name}")
        except Exception as e: print(f"  - Could not fetch or process {event_name}: {e}")
    if not real_data: raise ValueError("Could not download any real events. Check your internet connection.")
    return torch.stack(real_data), torch.stack(real_params)

# --- MAIN SCRIPT ---

cache_filename = 'ligo_synthetic_data_1000.pt'

if os.path.exists(cache_filename):
    print("Stage 1: Loading pre-generated synthetic data from cache...")
    x_train, y_train = torch.load(cache_filename)
else:
    start_time = time.time()

    x_train, y_train = generate_realistic_synthetic(1000)
    torch.save((x_train, y_train), cache_filename)
    print(f"Data generation and caching took {time.time() - start_time:.2f} seconds.")

train_dataset = TensorDataset(x_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)

model = LIGO_VAE()
optimizer = optim.Adam(model.parameters(), lr=5e-4)

print("\nStage 2: Pre-training on synthetic data...")
for epoch in range(801):
    model.train()
    total_loss = 0
    for batch_x, batch_y in train_loader:
        recon, pred_p, mu, log_var = model(batch_x)
        recon_loss = nn.functional.mse_loss(recon, batch_x, reduction='mean')
        reg_loss = nn.functional.mse_loss(pred_p, batch_y, reduction='mean') * 2.0
        kld_loss = -0.5 * torch.mean(1 + log_var - mu.pow(2) - log_var.exp())
        loss = recon_loss + reg_loss + (0.5 * kld_loss)
        optimizer.zero_grad(); loss.backward(); optimizer.step()
        total_loss += loss.item()

    if epoch % 200 == 0:
        print(f"Pre-training Epoch {epoch} | Avg Loss: {total_loss / len(train_loader):.4f}")

print("\nStage 3: Fine-tuning on real LIGO data...")
x_real, y_real = get_real_few_shot_dataset()
real_dataset = TensorDataset(x_real, y_real)
real_loader = DataLoader(real_dataset, batch_size=len(x_real))
fine_tune_optimizer = optim.Adam(model.parameters(), lr=1e-5)
for epoch in range(300):
    model.train()
    for batch_x, batch_y in real_loader:
        recon, pred_p, mu, log_var = model(batch_x)
        reg_loss = nn.functional.mse_loss(pred_p, batch_y, reduction='mean')
        kld_loss = -0.5 * torch.mean(1 + log_var - mu.pow(2) - log_var.exp())
        loss = reg_loss + (0.1 * kld_loss)
        fine_tune_optimizer.zero_grad(); loss.backward(); fine_tune_optimizer.step()
    if epoch % 100 == 0:
        with torch.no_grad():
            error = torch.mean(torch.abs(pred_p - batch_y)).item()
            print(f"Fine-tuning Epoch {epoch} | Avg Real-Data Error: {error:.2f}")

print("\nStage 4: Final validation on GW150914...")
t0 = event_gps('GW150914')
raw = TimeSeries.fetch_open_data('H1', t0-1, t0+1, cache=True)
clean = raw.whiten().bandpass(30, 400)
cropped = clean.crop(t0-0.02, t0+0.02).value[:100].copy()
test_input = torch.from_numpy(cropped).float()
test_input = (test_input - test_input.mean()) / test_input.std()
model.eval()
with torch.no_grad():
    _, prediction, _, _ = model(test_input.unsqueeze(0))

print("\n" + "="*40)
print(f"ROBUST FEW-SHOT LEARNING RESULTS")
print(f"AI Estimated TOTAL Mass: ~{prediction[0,0].item():.2f} Solar Masses")
print(f"Actual Known TOTAL Mass: ~65.3 Solar Masses")
print(f"AI Estimated Effective Spin (chi_eff): ~{prediction[0,1].item():.2f}")
print(f"Actual Known Effective Spin (chi_eff): ~-0.06")
print("="*40)
```
